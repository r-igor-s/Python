```python
def equation(x):
    return 2*x**4 - 8*x**3 - 16*x**2 - 1

def bisection_method(a, b, tol):
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if equation(c) == 0:
            return c
        elif equation(c) * equation(a) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

a = -2
b = 2

#Точність
tolerance = 0.001

root = bisection_method(a, b, tolerance)

#Результат
print(f"Рівняння має корінь при x = {root}")
#Перевірка
result = equation(root)
print(f"Значення функції при x = {root} дорівнює {result}")


def equation(x):
    return 2*x**4 + 4*x**3 + x**2 + 3*x - 6

def chord_method(x0, x1, tol, max_iter):
    iteration = 0
    while abs(equation(x1)) > tol and iteration < max_iter:
        x2 = x1 - (equation(x1) (x1 - x0)) / (equation(x1) - equation(x0))
        x0 = x1
        x1 = x2
        iteration += 1
    return x1

x0 = 0.5
x1 = 1.0

#Точність
tolerance = 0.001

#Максимальна кількість ітерацій
max_iterations = 1000

root = chord_method(x0, x1, tolerance, max_iterations)

#Результат
print(f"Рівняння має корінь при x = {root}")

#Перевірка
result = equation(root)
print(f"Значення функції при x = {root} дорівнює {result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

x = np.array([-2., -1., 0., 3.], dtype=float)
y = np.array([-4., 3., 2., 11.], dtype=float)
x_tests = [-1.5, -0.5, 1, 2]

def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n)
    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j]) / (x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p)

for x_test in x_tests:
    f_interp = lagrange_interpolation(x, y, x_test)
    print(f"Значення функції у точці x_test = {f_interp:.4f} для x = {x_test}")


xnew = np.linspace(np.min(x),np.max(x),100)
ynew = [lagrange_interpolation(x,y,i) for i in xnew] 

plt.plot(x,y,'o',xnew,ynew) 
plt.title('Lagrange Polynomial_1') 
plt.grid(True) 
plt.show()

#Перевірка 
f = lagrange(x, y) 
fig = plt.figure(figsize = (7,5)) 
plt.plot(xnew, f(xnew), 'b', x, y, 'ro') 
plt.title('Lagrange Polynomial_2') 
plt.grid() 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show()
```

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

x = np.array([0.1, 0.3, 0.6, 1.1, 1.8])
y = np.array([2.65, 2.75, 2.19, 1.76, 3.43])
n = len(x) - 1
h = np.diff(x)
a = y
b = np.zeros(n)
d = np.zeros(n)
c = np.zeros(n)
alpha = np.zeros(n)

for i in range(1, n):
    alpha[i] = (3 / h[i]) * (a[i+1] - a[i]) - (3 / h[i-1]) * (a[i] - a[i-1])

l = np.ones(n)
mu = np.zeros(n)
z = np.zeros(n)

for i in range(1, n):
    l[i] = 2 * (x[i+1] - x[i-1]) - h[i-1] * mu[i-1]
    mu[i] = h[i] / l[i]
    z[i] = (alpha[i] - h[i-1] * z[i-1]) / l[i]

c[n-1] = (alpha[n-1] - h[n-2] * z[n-2]) / (2 * (h[n-2] + mu[n-1]))
b[n-1] = (a[n] - a[n-1]) / h[n-1] - h[n-1] * (2 * c[n-1] + c[n-2]) / 3
d[n-1] = (c[n-1] - c[n-2]) / (3 * h[n-1])

for j in range(n - 2, -1, -1):
    c[j] = z[j] - mu[j] * c[j+1]
    b[j] = (a[j+1] - a[j]) / h[j] - h[j] * (c[j+1] + 2 * c[j]) / 3
    d[j] = (c[j+1] - c[j]) / (3 * h[j])
for i in range(n):
    print(f"Відрізок {i + 1}:")
    print(f"S_{i}(x) = {a[i]} + {b[i].round(4)}(x - {x[i]}) + {c[i].round(4)}(x - {x[i]})^2 + {d[i].round(4)}(x - {x[i]})^3, x належить [{x[i]}, {x[i+1]}]")

#Перевірка
cs = CubicSpline(x, y)
x_new = np.linspace(np.min(x), np.max(x), 100)
y_new = cs(x_new)
print("Значення сплайна:")
for i in range(len(x_new)):
    print(f"x = {x_new[i]:.2f}, y = {y_new[i]:.3f}")

#Побудова кубічного сплайна
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Точки')
plt.plot(x_new, y_new, label='Кубічний сплайн')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Інтерполяція кубічним сплайном')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares

def true_function(x):
    return np.cos(2 * x) + 2 * x

def linear_model(x, params):
    return params[0] * x + params[1]

def quadratic_model(x, params):
    return params[0] * x**2 + params[1] * x + params[2]

x_data = np.arange(0.1, 1.1, 0.1)
y_data = true_function(x_data)

def residual(params, x, y_true):
    return y_true - linear_model(x, params)

initial_params_linear = [1, 1]
result_linear = least_squares(residual, initial_params_linear, args=(x_data, y_data))

def residual_quadratic(params, x, y_true):
    return y_true - quadratic_model(x, params)

initial_params_quadratic = [1, 1, 1]
result_quadratic = least_squares(residual_quadratic, initial_params_quadratic, args=(x_data, y_data))

#Графік
x_range = np.linspace(0.05, 1.05, 100)
y_true = true_function(x_range)
y_linear = linear_model(x_range, result_linear.x)
y_quadratic = quadratic_model(x_range, result_quadratic.x)

plt.plot(x_range, y_true, label='Справжня функція')
plt.plot(x_range, y_linear, label='Лінійне наближення')
plt.plot(x_range, y_quadratic, label='Квадратичне наближення')
plt.title('Апроксимація методом найменших квадратів')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from scipy import integrate 
import numpy as np
from scipy import integrate

#Метод прямокутників, п=10
eps = 0.001
a = 0.8  
b = 1.4  

def f(x): 
    return 1 / (np.sqrt(2*x + 3))

def left_rec(f, a, b, n): 
    h = (b - a) / n 
    sum = 0 
    for i in range(0, n): 
        sum += f(a + i * h) 
    return sum * h 

def right_rec(f, a, b, n): 
    h = (b - a) / n 
    sum = 0 
    for i in range(1, n+1): 
        sum += f(a + i * h) 
    return sum * h 

def aver_rec(f, a, b, n): 
    h = (b - a) / n 
    sum = 0 
    for i in range(0, n): 
        sum += f(a + i * h) 
    return sum * h 

left_result = left_rec(f, a, b, 10)
print("Left rectangle:", round(left_result, 5)) 

right_result = right_rec(f, a, b, 10)
print("Right rectangle:", round(right_result, 5)) 

aver_result = aver_rec(f, a, b, 10)
print("Average rectangle:", round(aver_result, 5)) 

v, err = integrate.quad(f, a, b)
print("Check for the rectangle method =", round(v, 5))


#Метод Сімпсона, п=8;
def f(x):
    return np.sqrt(x) * np.cos(x**2)

a = 0.4
b = 1.2
n = 8

def simpson_rule(f, a, b, n):
    h = (b - a) / n 
    integr = f(a) + f(b) 
    for i in range(1, n): 
        k = a + i * h 
        if i % 2 == 0: 
            integr += 2 * f(k) 
        else: 
            integr += 4 * f(k) 
    integr *= h / 3 
    return integr 

integral1 = simpson_rule(f, a, b, n)
n *= 2
integral2 = simpson_rule(f, a, b, n)
while abs(integral2 - integral1) / 15 > 0.001:
    integral1 = integral2
    n *= 2
    integral2 = simpson_rule(f, a, b, n)

print("\nSimpson method:", round(integral2, 3))

v, err = integrate.quad(f, a, b)
print("Check for the Simpson method:", round(v, 3))


#Метод трапецій, п=20;
def f(x):
    return 1 / np.sqrt(3 * x**2 - 0.4)

a = 1.3
b = 2.1
n = 20

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    x = a
    sum = 0
    for i in range(1, n):
        x += h
        sum += 2 * f(x)
    sum += f(b)
    integral = h / 2 * sum
    return integral

integral1 = trapezoidal_rule(f, a, b, n)
n *= 2
integral2 = trapezoidal_rule(f, a, b, n)
while abs(integral2 - integral1) / 3 > 0.001:
    integral1 = integral2
    n *= 2
    integral2 = trapezoidal_rule(f, a, b, n)

print("\nTrapezoidal method:", round(integral2, 3))

#Перевірка
v, err = integrate.quad(f, a, b)
print("Check for the trapezoidal method:", round(v, 5))
```
